# Extremely Random Forests

```python
import numpy as np
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.multiclass import OneVsOneClassifier
from sklearn.multiclass import OneVsRestClassifier

# This provides a list of keys assiciated with each set of data.
# 0 - Charles Dickens
# 1 - HG Wells
# 2 - Jane Austen
# 3 - W Shakespeare
i = 0
keys = []
while i<68:
    keys.append(0)
    i = i+1
while i>=68 and i<130:
    keys.append(1)
    i = i+1
while i>=130 and i<206:
    keys.append(2)
    i = i+1
while i>=206 and i<259:
    keys.append(3)
    i = i+1

X_train, X_test, y_train, y_test = train_test_split(
    feature_set, keys, test_size=0.4, random_state=0)

# Extra trees
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import ExtraTreesClassifier

clf = ExtraTreesClassifier(n_estimators = 1000,
    max_depth = None, random_state = 1).fit(X_train, y_train)
bag_class = BaggingClassifier(clf, n_estimators = 4).fit(X_train, y_train)
clf_class_ovo = OneVsOneClassifier(bag_class).fit(X_train, y_train)
clf_class_ovr = OneVsRestClassifier(bag_class).fit(X_train, y_train)
clf_class_score_ovo = clf_class_ovo.score(X_test, y_test)
clf_class_score_ovr = clf_class_ovr.score(X_test, y_test)
score = bag_class.score(X_test,y_test)

score_percent = score*100
ovr_score_percent = clf_class_score_ovr*100
ovo_score_percent = clf_class_score_ovo*100

print('--EXTRA RANDOM FORESTS--')
print('This represents the percentage of tests the model correctly identifies.\n')
print('This is the score:')
print(str(score_percent) + "%")
print('This is the one vs rest score:')
print(str(ovr_score_percent) + "%")
print('This is the one vs one score:')
print(str(ovo_score_percent) + "%")
```
